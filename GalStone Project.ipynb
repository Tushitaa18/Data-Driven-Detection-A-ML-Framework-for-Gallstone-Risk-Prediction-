```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_fe = df.copy()

# ============================
# 1. CREATE NEW FEATURES
# ============================

# 1. HDL / LDL Ratio
df_fe['HDL_LDL_Ratio'] = df_fe['High Density Lipoprotein (HDL)'] / (df_fe['Low Density Lipoprotein (LDL)'] + 1e-5)

# 2. CRP / Vitamin D
df_fe['CRP_to_VitD'] = df_fe['C-Reactive Protein (CRP)'] / (df_fe['Vitamin D'] + 1e-5)

# 3. Fat Mass / Weight (%)
df_fe['FatMass_to_Weight'] = df_fe['Total Fat Content (TFC)'] / (df_fe['Weight'] + 1e-5)

# 4. ECW / ICW Ratio
df_fe['ECW_ICW_Ratio'] = df_fe['Extracellular Water (ECW)'] / (df_fe['Intracellular Water (ICW)'] + 1e-5)

# 5. AST / ALT Ratio (liver marker)
df_fe['AST_ALT_Ratio'] = df_fe['Aspartat Aminotransferaz (AST)'] / (df_fe['Alanin Aminotransferaz (ALT)'] + 1e-5)

# 6. High BMI flag (obese >= 30)
df_fe['High_BMI'] = (df_fe['Body Mass Index (BMI)'] >= 30).astype(int)

# ============================
# 2. DROP LOW-VARIANCE FEATURES
# ============================

low_var_cols = df_fe.columns[df_fe.nunique() == 1]
print("Dropping low-variance features:", list(low_var_cols))
df_fe.drop(columns=low_var_cols, inplace=True)

# ============================
# 3. DROP HIGHLY CORRELATED FEATURES
# ============================

# Compute correlation matrix
corr_matrix = df_fe.corr(numeric_only=True).abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Drop columns with correlation > 0.9
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
print("Dropping highly correlated features:", to_drop)
df_fe.drop(columns=to_drop, inplace=True)

# ============================
# 4. FINAL SHAPE AND CHECK
# ============================

print(f"Original shape: {df.shape}")
print(f"After Feature Engineering: {df_fe.shape}")

# OPTIONAL: Save cleaned dataset
# df_fe.to_csv("cleaned_feature_engineered_gallstone.csv", index=False)
```

## ✅ Summary of What Was Done

### 🔧 New Features Created (6)

These derived features were engineered to capture important clinical relationships:

- **HDL_LDL_Ratio** – balance of good (HDL) to bad (LDL) cholesterol  
- **CRP_to_VitD** – inflammation vs. Vitamin D status  
- **FatMass_to_Weight** – proportion of fat in overall weight  
- **ECW_ICW_Ratio** – ratio of extracellular to intracellular water  
- **AST_ALT_Ratio** – marker of liver health  
- **High_BMI** – binary indicator for obesity (BMI ≥ 30)

---

### 🗑️ Features Removed

#### 🔹 Low variance features:  
- **None found** — ✅ Good!

#### 🔹 Highly correlated features (correlation coefficient > 0.9):  
To avoid multicollinearity and redundancy, the following were dropped:

- `Extracellular Water (ECW)`  
- `Lean Mass (LM) (%)`  
- `Muscle Mass (MM)`  
- `Total Fat Content (TFC)`  
- `FatMass_to_Weight` *(derived feature)*  
- `ECW_ICW_Ratio` *(derived feature)*

---

Removing these ensures model robustness and better generalization by reducing feature redundancy.

## Unsupervised clustering algorithm

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# ==== Step 1: Preprocess Data ====
# Assuming 'df' is your main DataFrame
X = df.drop(columns=['Gallstone Status Label', 'Gallstone Status'])
y = df['Gallstone Status Label']

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ==== Step 2: Dimensionality Reduction (optional but improves KMeans) ====
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# ==== Step 3: Fit KMeans ====
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(X_pca)

# ==== Step 4: Map KMeans Labels to True Labels ====
# Try both permutations to account for arbitrary label assignment
mapped_labels_1 = cluster_labels
mapped_labels_2 = 1 - cluster_labels

acc1 = accuracy_score(y, mapped_labels_1)
acc2 = accuracy_score(y, mapped_labels_2)

# Choose the better mapping
if acc1 > acc2:
    final_preds = mapped_labels_1
else:
    final_preds = mapped_labels_2

# ==== Step 5: Evaluation ====
print("🔹 K-Means Clustering (with PCA)")
print("Accuracy:", accuracy_score(y, final_preds))
print(classification_report(y, final_preds))

# ==== Step 6: Optional Cluster Visualization ====
plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=final_preds, cmap='coolwarm', edgecolors='k', s=60)
plt.title("K-Means Clustering Visualization (PCA-reduced space)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.savefig("K-Mean.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Install PyTorch if not installed
try:
    import torch
except ImportError:
    !pip install torch
    import torch

import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import classification_report
import numpy as np

# Convert data to torch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)

# Create DataLoader for batch training
batch_size = 16
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Define improved neural network
class ImprovedGallstoneNN(nn.Module):
    def __init__(self, input_dim):
        super(ImprovedGallstoneNN, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.LayerNorm(128),
            nn.LeakyReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, 64),
            nn.LayerNorm(64),
            nn.LeakyReLU(),
            nn.Dropout(0.3),

            nn.Linear(64, 32),
            nn.LayerNorm(32),
            nn.LeakyReLU(),
            nn.Dropout(0.2),

            nn.Linear(32, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.network(x)

# Instantiate model
model = ImprovedGallstoneNN(input_dim=X_train.shape[1])
criterion = nn.BCELoss()
optimizer = optim.AdamW(model.parameters(), lr=0.0007, weight_decay=1e-4)

# Training with Early Stopping
epochs = 150
best_loss = np.inf
early_stop_counter = 0
patience = 15

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        output = model(X_batch)
        loss = criterion(output, y_batch)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    # Validation
    model.eval()
    with torch.no_grad():
        val_output = model(X_test_tensor)
        val_loss = criterion(val_output, y_test_tensor)

    print(f"Epoch {epoch+1}: Train Loss = {running_loss:.4f}, Val Loss = {val_loss.item():.4f}")

    if val_loss.item() < best_loss:
        best_loss = val_loss.item()
        best_model = model.state_dict()
        early_stop_counter = 0
    else:
        early_stop_counter += 1

    if early_stop_counter >= patience:
        print(f"⛔ Early stopping triggered at epoch {epoch+1}")
        break

# Load best model
model.load_state_dict(best_model)

# Evaluation
model.eval()
with torch.no_grad():
    y_pred_probs = model(X_test_tensor).numpy()
    y_pred = (y_pred_probs > 0.5).astype(int)
    print("\n✅ Improved Custom Neural Network Evaluation:\n")
    print(classification_report(y_test, y_pred, digits=4))
```
