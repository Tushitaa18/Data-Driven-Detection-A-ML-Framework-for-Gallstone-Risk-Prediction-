```python
# 1. Correlation matrix (already plotted before, but shown here for completeness)
plt.figure(figsize=(18, 16))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm', square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Gallstone Dataset')
plt.tight_layout()
plt.show()

# 2. Top 5 features most correlated with Gallstone Status
target_corr = corr['Gallstone Status'].drop('Gallstone Status').sort_values(key=abs, ascending=False)
print("Top 5 correlated features with Gallstone Status:\n", target_corr.head())

# 3. Pairplot with top 3 correlated features
top_features = target_corr.head(3).index.tolist()
sns.pairplot(df, vars=top_features + ['Gallstone Status'], hue='Gallstone Status', palette='Set2')
plt.suptitle('Pairplot of Top Correlated Features with Gallstone Status', y=1.02)
plt.show()

# 4. Scatter plots of top features vs Gallstone Status
for col in top_features:
    sns.stripplot(x='Gallstone Status', y=col, data=df, jitter=True, palette='coolwarm')
    plt.title(f'{col} vs Gallstone Status')
    plt.show()
```

```python
#Feature Engineering
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_fe = df.copy()

# ============================
# 1. CREATE NEW FEATURES
# ============================

# 1. HDL / LDL Ratio
df_fe['HDL_LDL_Ratio'] = df_fe['High Density Lipoprotein (HDL)'] / (df_fe['Low Density Lipoprotein (LDL)'] + 1e-5)

# 2. CRP / Vitamin D
df_fe['CRP_to_VitD'] = df_fe['C-Reactive Protein (CRP)'] / (df_fe['Vitamin D'] + 1e-5)

# 3. Fat Mass / Weight (%)
df_fe['FatMass_to_Weight'] = df_fe['Total Fat Content (TFC)'] / (df_fe['Weight'] + 1e-5)

# 4. ECW / ICW Ratio
df_fe['ECW_ICW_Ratio'] = df_fe['Extracellular Water (ECW)'] / (df_fe['Intracellular Water (ICW)'] + 1e-5)

# 5. AST / ALT Ratio (liver marker)
df_fe['AST_ALT_Ratio'] = df_fe['Aspartat Aminotransferaz (AST)'] / (df_fe['Alanin Aminotransferaz (ALT)'] + 1e-5)

# 6. High BMI flag (obese >= 30)
df_fe['High_BMI'] = (df_fe['Body Mass Index (BMI)'] >= 30).astype(int)

# ============================
# 2. DROP LOW-VARIANCE FEATURES
# ============================

low_var_cols = df_fe.columns[df_fe.nunique() == 1]
print("Dropping low-variance features:", list(low_var_cols))
df_fe.drop(columns=low_var_cols, inplace=True)

# ============================
# 3. DROP HIGHLY CORRELATED FEATURES
# ============================

# Compute correlation matrix
corr_matrix = df_fe.corr(numeric_only=True).abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Drop columns with correlation > 0.9
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
print("Dropping highly correlated features:", to_drop)
df_fe.drop(columns=to_drop, inplace=True)

# ============================
# 4. FINAL SHAPE AND CHECK
# ============================

print(f"Original shape: {df.shape}")
print(f"After Feature Engineering: {df_fe.shape}")

# OPTIONAL: Save cleaned dataset
# df_fe.to_csv("cleaned_feature_engineered_gallstone.csv", index=False)
```

## ✅ Summary of What Was Done

### 🔧 New Features Created (6)

These derived features were engineered to capture important clinical relationships:

- **HDL_LDL_Ratio** – balance of good (HDL) to bad (LDL) cholesterol  
- **CRP_to_VitD** – inflammation vs. Vitamin D status  
- **FatMass_to_Weight** – proportion of fat in overall weight  
- **ECW_ICW_Ratio** – ratio of extracellular to intracellular water  
- **AST_ALT_Ratio** – marker of liver health  
- **High_BMI** – binary indicator for obesity (BMI ≥ 30)

---

### 🗑️ Features Removed

#### 🔹 Low variance features:  
- **None found** — ✅ Good!

#### 🔹 Highly correlated features (correlation coefficient > 0.9):  
To avoid multicollinearity and redundancy, the following were dropped:

- `Extracellular Water (ECW)`  
- `Lean Mass (LM) (%)`  
- `Muscle Mass (MM)`  
- `Total Fat Content (TFC)`  
- `FatMass_to_Weight` *(derived feature)*  
- `ECW_ICW_Ratio` *(derived feature)*

---

Removing these ensures model robustness and better generalization by reducing feature redundancy.
